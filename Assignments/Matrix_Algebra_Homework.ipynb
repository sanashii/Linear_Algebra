1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.

```python
def matrixAddition(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.") # used raise for exception handling
    
    # result matrix is an empty matrix with the same dimensions as the passed matrices
    rows, cols = len(matrix1), len(matrix1[0])
    result = [[0 for _ in range(cols)] for _ in range(rows)]
    

    for i in range(rows):
        for j in range(cols):
            result[i][j] = matrix1[i][j] + matrix2[i][j] # matrix addition
    
    return result

# sample application of function
matrix1 = [[12, 2, 7], 
           [4, 17, 1]]
matrix2 = [[5, 8, 0], 
           [11, 50, 6]]

result = matrixAddition(matrix1, matrix2)
for row in result:
        print(row)
```

2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrixMultiplication(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Number of columns in the first matrix must be equal to the number of rows in the second matrix.") # used raise for exception handling

    # Determine the dimensions of the result matrix
    rows1, cols1 = len(matrix1), len(matrix1[0])
    rows2, cols2 = len(matrix2), len(matrix2[0])
    
    # Initialize an empty result matrix
    result = [[0 for _ in range(cols2)] for _ in range(rows1)]

    # Perform matrix multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

# sample application of function
matrix1 = [[12, 2], 
           [4, 17]]
matrix2 = [[5, 8], 
           [11, 6]]

result = matrixMultiplication(matrix1, matrix2)
for row in result:
        print(row)
```

4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extractDia(matrix):
    rows, cols = len(matrix), len(matrix[0])
    if rows != cols:
        raise ValueError("Matrix must be square to have a diagonal.")

    diagonal = [matrix[i][i] for i in range(rows)]
    return diagonal

# sample application of function
matrix = [[1, 2, 3], 
          [4, 5, 6], 
          [7, 8, 9]]

diagonal = extractDia(matrix)
print("Diagonal elements:", diagonal)
```

5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrixTrace(matrix):
    rows, cols = len(matrix), len(matrix[0])
    if rows != cols:
        raise ValueError("Matrix must be square to have a trace.")

    trace = 0
    for i in range(rows):
        trace += matrix[i][i]

    return trace

# sample application of function
matrix = [[2, 4, 6], 
          [8, 10, 12], 
          [14, 16, 18]]

trace = matrixTrace(matrix)
print("Trace of the matrix:", trace)
```
